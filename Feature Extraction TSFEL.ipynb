```python
import numpy as np
import tsfel
import matplotlib.pyplot as plt
import json

train_sig = np.load("./data/train_without_resp.npy")
test_sig = np.load("./data/test_without_resp.npy")
train_features = np.load("./data/features_train_without_resp.npy")
test_features = np.load("./data/features_test_without_resp.npy")

with open('./data/train_signal_template.json', 'r') as file:
    train_sig_template = json.load(file)
with open('./data/test_signal_template.json', 'r') as file:
    test_sig_template = json.load(file)
with open('./json/all_features.json', 'r') as file:
    cfg_file = json.load(file)
```

```python
print(train_sig.shape)
print(test_sig.shape)
print(len(train_sig_template))
print(len(test_sig_template))
print(train_features.shape)
print(test_features.shape)
print(len(train_sig_template[1]))
```

```python
# tmp = train_sig_template[0]

# cfg_file = tsfel.get_features_by_domain()
# cgf_file = tsfel.get_features_by_domain("statistical")  # All statistical domain features will be extracted
# cgf_file = tsfel.get_features_by_domain("temporal")     # All temporal domain features will be extracted
# cgf_file = tsfel.get_features_by_domain("spectral")
# If no argument is passed retrieves all available features

train_new_features = np.zeros(67)

for i in range(len(train_sig_template)):
    features = tsfel.time_series_features_extractor(cfg_file, train_sig_template[i], fs=100, window_size=len(train_sig_template[i])).values.flatten()
    # Receives a time series sampled at 50 Hz, divides into windows of size 250 (i.e. 5 seconds) and extracts all features
    new_features = np.concatenate((features, train_features[i, :]))
    
    train_new_features = np.vstack((train_new_features, new_features))
    
train_new_features = train_new_features[1:, :]
print(train_new_features.shape)
print(train_new_features[55, :])
print(train_features[55, :])
```
